```python
# analyze data by building a network of co-occuring tags and running a Lasso regression to find top positive/negative tags for recipe ratings

import sys
import os
repo_root = os.path.abspath("..")
sys.path.append(os.path.join(repo_root, "src"))

from data_utils import RecipeAnalysis

data_path = os.path.join(repo_root, "data", "raw", "epi_r.csv")
ra = RecipeAnalysis(data_path)
ra.preprocess()

# build cooccurence graph
cooc_graph = ra.build_coocurrence_graph(threshold=100)
print("Number of nodes in co-occurrence graph:", len(cooc_graph.nodes()))
print("Number of edges in co-occurrence graph:", len(cooc_graph.edges()))

# run lasso
lasso_df = ra.run_lasso(target_col='rating')
print("Top 10 positive tags:")
print(lasso_df.head(10))
print("\nTop 10 negative tags:")
print(lasso_df.tail(10))
```
